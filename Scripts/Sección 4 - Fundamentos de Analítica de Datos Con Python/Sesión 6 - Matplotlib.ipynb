### Matplotlib 
- Librería para construir visualizaciones de datos. 
    - Diagramas de barras
    - Histogramas 
    - Diagramas circulares (pastel) 
    - Diagrama de caja y bigotes 
    - Diagrama de dispersión 
    - Lineas 

##### Creando visualización con MATPLOTLIB 
##### Pyplot 

- Importar pyplot 
- Definir la figura o figuras 
- Dibujar los datos sobre los ejes 
- Personaliza 
- Guarda o se muestra 

```python
## Carga inicial 
import pandas as pd 
import matplotlib.pyplot as plt 

videojuegos = pd.read_excel('videojuegos.xlsx')
ozone = pd.read_csv('ozone.csv')
```

```python
# creamos la figura y los ejes 
fig, ax = plt.subplots()
# dibujamos los puntos 
ax.scatter(x = [1, 2 ,3], y=[3,2,1])
# mostrando la figura 
plt.show()
```

```python
ozone.info()
```

```python
## Scatterplot o dispersión - gráfico que va a dibujar puntos en los pares (x,y)

plt.scatter(ozone['Humidity'], ozone['Temperature_ElMonte'])
```

```python
# Diagrama de lineas 
plt.plot([1, 2 ,3 ,1 ,2,3])
```

```python
# diagrama de áreas 
plt.fill_between([1, 2 ,3 ,4], [1, 3 ,0.5 ,5])
```

```python
# Diagrama de barras 
plt.bar(['Lunes', 'Martes', 'Miercoles'], [3, 6 ,9])
```

```python
# diagrama de barras horizontales
plt.barh(['Lunes', 'Martes', 'Miercoles'], [3, 6 ,9])
```

```python
# histograma 
import numpy as np
x = np.random.normal(5, 1.5, 100)

plt.hist(x)
```

```python
plt.pie([5, 6 ,4 ,8 ,5]);
```

```python
# diagrama de caja y bigotes 
x = np.random.normal(5, 1.5, 100);

plt.boxplot(x);
```

```python
# Diagrama de violín

plt.violinplot(x);
```

```python
x = np.random.random((5,5))
plt.imshow(x)
```

```python
### Colores, marcadores, estilos de linea, titulos, ejes, leeyendas, grid 
```

```python
#colores, hexadeximal, Cn, nombre, o en rgb 

plt.plot([1, 2 ,3 ,1 ,2,3], color = 'BlueViolet');
```

```python
plt.plot([1, 2 ,3 ,1 ,2,3], 
         color = 'BlueViolet', 
        marker = 'v',
        linestyle = ':');
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2 ,3 ,1 ,2,3], 
         color = 'BlueViolet', 
        marker = 'v',
        linestyle = ':');
ax.set_title('Ejemplo de Gráfico', loc='Right', fontdict ={
  'color': 'violet',
    'fontsize':15}
            );
```

```python
ax.plot([1, 8,-2 ,1 ,6,3], 
         color = 'Crimson', 
        marker = 'v',
        linestyle = ':');


# ejes 
# labels de los ejes 
ax.set_xlabel('Pasos');
ax.set_ylabel('Niveles')
ax.set_xlim([-2,6])
ax.set_ylim([-4,9])
# scalas, limites
fig
```

```python
ax.legend(['Grafico1', 'Gráfico 2'], loc = 'upper left')
fig
```

```python
ax.grid(axis = 'x', color = '#0e0e0e', linestyle = 'dashed')
fig
```

```python
# Elimino el contenido de la figura guardada en memoria 
fig.clear()
```
